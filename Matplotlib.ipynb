```python
# Matplotlib

import matplotlib.pyplot as plt
import numpy as np

# 1. Gráfico de líneas

dias = ['Lun', 'Mar', 'Mie', 'Jue', 'Vie', 'Sab', 'Dom']
temp = [22, 24, 21, 25, 28, 26, 23]

# Estilo del gráfico
plt.plot(dias, temp, color='green', marker='o', linestyle='--')

plt.title('Cambio de temperatura en la semana')
plt.xlabel('Día')
plt.ylabel('Grados Celsius')
plt.grid(True)
plt.show()
```

```python
# 2. Gráfico de barras

productos = ['Pizza', 'Hamburguesa', 'Arepa', 'Salchipapa']
ventas = [15, 22, 18, 10]

plt.bar(productos, ventas, color='blue')

plt.title('Comidas más vendidas hoy')
plt.xlabel('Producto')
plt.ylabel('Unidades vendidas')
plt.show()
```

```python
# 3. Gráfico de dispersión

# X: Horas dormidas, Y: Nivel de energía (1 al 10)
horas_sueno = [5, 6, 7, 8, 4, 9, 7.5]
nivel_energia = [4, 6, 8, 9, 3, 10, 7]

plt.scatter(horas_sueno, nivel_energia, color='purple')

plt.title('Relación entre sueño y energía')
plt.xlabel('Horas de sueño')
plt.ylabel('Nivel de energía')
plt.show()
```

```python
# 4. Gráfico de pastel o torta

# Actividades y el porcentaje de tiempo que ocupan
actividades = ['Estudiar', 'Dormir', 'Trabajar', 'Comer']
tiempo = [30, 35, 25, 10]
colores = ['purple', 'lightblue', 'lightpink', 'cyan']

plt.pie(tiempo, labels=actividades, autopct='%1.1f%%', colors=colores, startangle=90)

plt.title('Distribución de mi día')
plt.show()
```

```python
# 5. Histograma

# Edades de un grupo de personas en un curso
edades = [18, 19, 21, 18, 20, 22, 19, 18, 25, 28, 19, 20, 18, 21, 20]

plt.hist(edades, bins=5, color='brown', edgecolor='black')

plt.title('Distribución de edades en el salón')
plt.xlabel('Rango de edad')
plt.ylabel('Número de personas')
plt.show()
```
