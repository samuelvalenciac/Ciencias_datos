```python
# 1. Abstracción
class Computador:
  def __init__(self, marca, procesador, ram):
    self.marca = marca
    self.procesador = procesador
    self.ram = ram

  def encender(self):
    print(f"El computador {self.marca} se está iniciando...")

# Creamos una instancia (objeto)
pc = Computador("Asus", "Intel i7", "16GB")
pc.encender()
```

```python
# 2. Encapsulamiento
class CuentaBanco:
  def __init__(self, titular, saldo_inicial):
    self.__titular = titular
    self.__saldo = saldo_inicial

  def ver_saldo(self):
    return f"Hola {self.__titular}, tu saldo actual es: ${self.__saldo}"

  def depositar(self, cantidad):
    if cantidad > 0:
      self.__saldo += cantidad
      print(f"Depósito exitoso. Nuevo saldo: ${self.__saldo}")
    else:
      print("No puedes depositar menos de 0")

cuenta = CuentaBanco("Samuel", 500)
print(cuenta.ver_saldo())
cuenta.depositar(200)
```

```python
# 3. Herencia

class Persona:
  def __init__(self, nombre, edad):
    self.nombre = nombre
    self.edad = edad

  def presentarse(self):
    print(f"Hola, soy {self.nombre} y tengo {self.edad} años.")

# Nueva clase hija de Persona
class Estudiante(Persona):
  def __init__(self, nombre, edad, carrera):
    super().__init__(nombre, edad)
    self.carrera = carrera

  def estudiar(self):
    print(f"{self.nombre} está estudiando la carrera de {self.carrera}.")

alumno1 = Estudiante("Samuel Valencia", 20, "Química farmacéutica")
alumno1.presentarse()
alumno1.estudiar()
```

```python
# 4. Polimorfismo

class Transporte:
  def __init__(self, nombre):
    self.nombre = nombre

  def avanzar(self, forma):
    print(f"El {self.nombre} está avanzando {forma}")

class Carro(Transporte):
  def __init__(self, nombre):
    super().__init__(nombre)

class Lancha(Transporte):
  def __init__(self, nombre):
    super().__init__(nombre)

carro1 = Carro("Nissan")
lancha1 = Lancha("Yate")

carro1.avanzar("por la carretera")
lancha1.avanzar("por el agua")
```

```python
# Principios SOLID

# 5. SOLID - Single Responsability

# Clase con una responsabilidad (datos tarea)
class Tarea:
  def __init__(self, titulo, descripcion):
    self.titulo = titulo
    self.descripcion = descripcion

# Clase de visualización
class GestorVisual:
  def mostrar_en_pantalla(self, tarea):
    print(f"MOSTRANDO TAREA:\nTítulo: {tarea.titulo}\nDetalle: {tarea.descripcion}")

# Uso
mi_tarea = Tarea("Aprender a usar POO", "Hacer 5 ejercicios con POO en Google colab y subirlo a un repositorio de GitHub")
visualizador = GestorVisual()
visualizador.mostrar_en_pantalla(mi_tarea)
```
